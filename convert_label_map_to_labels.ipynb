## Load Label Map from File

```python
from object_detection.utils import label_map_util
```

```python
label_map_path = "./input/label_map.pbtxt"
```

```python
label_map = label_map_util.load_labelmap(label_map_path)
```

```python
label_map
```

```python
max_label_map_index = label_map_util.get_max_label_map_index(label_map)
max_label_map_index
```

```python
max_label_map_index = 90
```

```python
categories = label_map_util.convert_label_map_to_categories(
    label_map, max_label_map_index)
categories
```

```python
category_index = label_map_util.create_category_index(categories)
category_index
```

## Extract Label Names without Background

```python
labels = label_map_util.extract_label_names_from_category_index(
    category_index, max_label_map_index)
labels
```

```python
labels_path = "./output/labels.txt"
```

```python
contents = '\n'.join(labels)
contents
```

```python
tf.io.write_file(labels_path, contents)
```

```python
!cat output/labels.txt
```

## Extract Label Names with Background

```python
labels_w_background = label_map_util.extract_label_names_from_category_index(
    category_index, max_label_map_index, fill_in_background=True)
labels_w_background
```

```python
labels_w_background_path = "./output/labels_w_background.txt"
```

```python
contents = '\n'.join(labels_w_background)
contents
```

```python
tf.io.write_file(labels_w_background_path, contents)
```
